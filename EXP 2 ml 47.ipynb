```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
import seaborn as sns 

#load titanic dataset 
titanic_data = sns.load_dataset('titanic')
```

```python
# normalization
titanic_df = sns.load_dataset('titanic')


titanic_df['age'] = (titanic_df['age'] - titanic_df['age'].min()) / (titanic_df['age'].max() - titanic_df['age'].min())


print(titanic_df['age'])
```

```python
# standardization
titanic_df['fare'] = (titanic_df['fare'] - titanic_df['fare'].mean()) / titanic_df['fare'].std()


print(titanic_df['fare'])
```

```python
print(titanic_data.describe())
```

```python
missing_values=titanic_data.isnull().sum()
print(missing_values)
```

```python
new_titanic_df=titanic_data.drop(columns=['deck'])
```

```python
new_titanic_df['age'].fillna(new_titanic_df['age'].median (),inplace=True)
new_titanic_df
```

```python
missing_values=new_titanic_df.isnull().sum()
print(missing_values)
```

```python
data=new_titanic_df
data['embark_town'].dtype
data['embark_town'].unique()
data['embark_town'].fillna(data['embark_town'].mode()[0],inplace=True)
data.isnull().sum()
```

```python
data=new_titanic_df
data['embarked'].dtype
data['embarked'].unique()
data['embarked'].fillna(data['embarked'].mode()[0],inplace=True)
data.isnull().sum()
```

```python
le=LabelEncoder()
data['sex']=le.fit_transform(data['sex'])
data['embarked']=le.fit_transform(data['embarked'])
```

```python
data.head()
```

```python
data=data[['pclass','sex','age','fare','embarked','survived']]

x=data[['pclass','sex','age','fare','embarked']]
y=data[['survived']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.3, random_state=42)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
LogisticRegression(max_iter=1000)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy",accuracy)
```
